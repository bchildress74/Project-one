```python
#read csv files

#financial data
financialsCSV = "Brett_Resources/finance_states.csv" 
finance_data = pd.read_csv(financialsCSV) 

#SAT scores 
SAT_CSV = "Brett_Resources/SAT_scores.csv" 
SAT_data = pd.read_csv(SAT_CSV) 

#grad rates 
grad_ratescsv = "Brett_Resources/grad_rates.csv" 
grad_rates = pd.read_csv(grad_ratescsv)

#crime rates
file='Resources - Christin/estimated_crimes.csv'
crime_rates = pd.read_csv(file)
```

```python
#Looking at 2015 SAT Scores compared to 2015 Crime Rates
#loading file 
SAT_data = "Brett_Resources/SAT_scores.csv"

#Reading the files 
SAT_df = pd.read_csv(SAT_data)

SAT_df.head()
```

```python
#Dropping unnecessary columns
SAT_df = SAT_df.drop(['EBRW', 'Math', 'Participation Rate'], axis=1)
SAT_df.head()
```

```python
Crime_data ="Resources - Christin/estimated_crimes.csv"

#Reading the files 
Crime_df = pd.read_csv(Crime_data)

Crime_df['Total Crime']=Crime_df['violent_crime']+Crime_df['homicide']+Crime_df['rape_legacy']+Crime_df['rape_revised']+Crime_df['robbery']\
+Crime_df['aggravated_assault']+Crime_df['property_crime']+Crime_df['burglary']+Crime_df['larceny']+Crime_df['motor_vehicle_theft']

year_2015 = Crime_df['YEAR']==2015
data_2015 = Crime_df[year_2015]   

data_2015 = data_2015.drop(['population','caveats','violent_crime'], axis=1)

fin_crimedata_2015 = data_2015.drop(['homicide','rape_legacy','rape_revised','robbery','aggravated_assault',\
                            'property_crime','burglary','larceny','motor_vehicle_theft','YEAR'], axis=1)
fin_crimedata_2015.head()
```
