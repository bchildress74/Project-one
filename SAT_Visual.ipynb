```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
#loading file 
SAT_data = "Brett_Resources/SAT_scores.csv"

#Reading the files 
SAT_df = pd.read_csv(SAT_data)

SAT_df.head()
```

```python
SAT_df = SAT_df.drop(['EBRW', 'Math', 'Participation Rate'], axis=1)
SAT_df.head()
```

```python
x_axis = np.arange(len(SAT_df))
tick_locations = [value+0.4 for value in x_axis]
```

```python
# Create a bar chart based off of the group series from before
plt.figure(figsize=(30,15))
plt.bar(SAT_df["State"], SAT_df["Total"], alpha=0.5, align="edge")
plt.xticks(tick_locations, SAT_df["State"], rotation="90", fontsize=20)
plt.yticks(fontname="Arial", fontsize=20)
plt.title("Total SAT Score by State", fontname="Arial", fontsize=30)
plt.xlabel("State", fontname="Arial", fontsize=30)
plt.ylabel("Total Score", fontname="Arial", fontsize=30)
plt.show()

```
